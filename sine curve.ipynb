```python
import numpy as np

def sigmoid(x):
    return 1/(1 + np.exp(-x/100))


def sigmoid_derivative(x):
    return x * (1 - x)

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

x = np.linspace(0, 3.14, 1000)
y = np.sin(x)
dataset = np.column_stack((x, y))

X = dataset[:,0]
y = dataset[:,1]

X = X.reshape(-1, 1)
y = y.reshape(-1, 1)

class NeuralNetwork:
    def __init__(self, x, y):
        self.input = x
        self.weights1 = np.random.rand(self.input.shape[1], 4) #  no of colomns in input and no of nodes in hidden layer
        self.weights2 = np.random.rand(4, 1)  #  no of nodes in hidden layer and no of colomns in output layer
        self.y = y
        self.output = np.zeros(self.y.shape)

    def feedforward(self):
        self.layer1 = relu(np.dot(self.input, self.weights1))
        self.output = sigmoid(np.dot(self.layer1, self.weights2))

    def backprop(self):
        d_weights2 = np.dot(self.layer1.T, (2*(self.y - self.output) * sigmoid_derivative(self.output)))
        d_weights1 = np.dot(self.input.T,  (np.dot(2*(self.y - self.output) * sigmoid_derivative(self.output), self.weights2.T) * relu_derivative(self.layer1)))

        self.weights1 += d_weights1
        self.weights2 += d_weights2

    def train(self, X, y):
        self.output = np.zeros(self.y.shape)
        self.input = X
        self.y = y
        self.feedforward()
        self.backprop()
        
nn = NeuralNetwork(X, y)
for i in range(1500):
    nn.train(X, y)
```

```python
print(nn.output)
```

```python
import matplotlib.pyplot as plt

plt.plot(X, y, label='Target')
plt.plot(X, nn.output, label='Output')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
def mse(y, y_pred):
    return np.mean((y - y_pred)**2)

loss = mse(y, nn.output)
print("Loss:", loss)
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.activations import relu, sigmoid

# generate some data
x = np.linspace(0, 3.14, 1000)
y = np.sin(x) * np.cos(x)
dataset = np.column_stack((x, y))

# split the data into training and testing sets
split_index = int(len(dataset) * 0.8)
train_data = dataset[:split_index, :]
test_data = dataset[split_index:, :]

# split the training and testing data into inputs and outputs
train_X = train_data[:, 0]
train_y = train_data[:, 1]
test_X = test_data[:, 0]
test_y = test_data[:, 1]

# define the model architecture
model = Sequential()
model.add(Dense(8, input_dim=1, activation=relu))
model.add(Dense(8, activation=relu))
model.add(Dense(1, activation=sigmoid))

# compile the model with a suitable loss function and optimizer
model.compile(loss='mean_squared_error', optimizer='adam')

# train the model on the training data
history = model.fit(train_X, train_y, epochs=1000, validation_data=(test_X, test_y))

# evaluate the model on the test data
loss = model.evaluate(test_X, test_y)

# print the final loss
print('Test Loss:', loss)
```

```python
import matplotlib.pyplot as plt

# get the predicted values for the test inputs
predicted_y = model.predict(test_X)

# plot the predicted values against the actual values
plt.scatter(test_X, test_y, label='Actual')
plt.scatter(test_X, predicted_y, label='Predicted')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from keras.models import Sequential
from keras.layers import Dense
from keras.activations import relu, sigmoid

x = np.linspace(0, 3.14, 1000)
y = np.sin(x)
dataset = np.column_stack((x, y))

X = dataset[:, 0]
y = dataset[:, 1]

model = Sequential()
model.add(Dense(32, input_dim=1, activation=relu))
model.add(Dense(32, activation=relu))
model.add(Dense(1, activation=sigmoid))
model.compile(loss='mse', optimizer='adam')

model.fit(X, y, epochs=1500, verbose=1)

# Get the predicted values
y_pred = model.predict(X).squeeze()

# Calculate R2 score
r2 = r2_score(y, y_pred)

# Plot the predicted values against the targets
plt.plot(X, y, label='target')
plt.plot(X, y_pred, label='predicted')
plt.legend()
plt.show()

print("R2 Score:", r2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
from keras.models import Sequential
from keras.layers import Dense
from keras.activations import relu, sigmoid

x = np.linspace(0, 3.14, 1000)
y = np.sin(x)
dataset = np.column_stack((x, y))

X = dataset[:, 0]
y = dataset[:, 1]

model = Sequential()
model.add(Dense(32, input_dim=1, activation=relu))
model.add(Dense(32, activation=relu))
model.add(Dense(1, activation=sigmoid))
model.compile(loss='mse', optimizer='adam')

model.fit(X, y, epochs=1500, verbose=1)

# Get the predicted values
y_pred = model.predict(X).squeeze()

# Calculate R2 score and MSE
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

# Plot the predicted values against the targets
plt.plot(X, y, label='target')
plt.plot(X, y_pred, label='predicted')
plt.legend()
plt.show()

print("MSE:", mse)
print("R2 Score:", r2)
```
